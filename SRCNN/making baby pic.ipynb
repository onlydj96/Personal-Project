```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

원본 이미지 로드 및 전처리

```python
img = cv2.imread('baby.jpg')
# img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def modcrop(img, scale):
    tmpsz = img.shape
    sz = tmpsz[0:2]
    sz = sz - np.mod(sz, scale)
    img = img[0:sz[0], 0:sz[1]]
    return img

img = modcrop(img, 2)

def shave(image, border):
    img = image[border:-border, border:-border]
    return img

img.shape
```

```python
factor = 2
h, w, c = img.shape
new_width = int(w / factor)
new_height = int(h / factor)

img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_LINEAR)  # downgraded
# downgraded_img = cv2.resize(img, (w, h), interpolation=cv2.INTER_LINEAR)  # downgraded한 것을 원래의 w, h로 복원

# 이미지 저장
# cv2.imwrite('baby.jpg', img)
img.shape
```

downgraded

```python
factor = 2
h, w, c = img.shape
new_width = int(w / factor)
new_height = int(h / factor)

downgraded_img = cv2.resize(img, (new_width, new_height), interpolation=cv2.INTER_LINEAR)  # downgraded
downgraded_img = cv2.resize(downgraded_img, (w, h), interpolation=cv2.INTER_LINEAR)  # downgraded한 것을 원래의 w, h로 복원


# downgraded_img = cv2.cvtColor(downgraded_img, cv2.COLOR_RGB2BGR)
downgraded_img.shape
```

```python
cv2.imwrite('downgraded_img.jpg', downgraded_img)
```

원본과 비교

```python
fig = plt.figure()
rows = 1; cols = 2

ax1 = fig.add_subplot(rows, cols, 1)
ax1.set_title('Original image')
ax1.imshow(img)

ax2 = fig.add_subplot(rows, cols, 2)
ax2.set_title('Downgraded image')
ax2.imshow(downgraded_img)

fig.tight_layout()
plt.show()
```
